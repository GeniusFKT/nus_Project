# 存在替代关系A=B时，合并两列

```python
def merge_col(df,a,b):
    mask=df[b].isnull()
    mask.head()
    df[b][mask]=df[a][mask]
    df=df.drop([a],axis=1)
    return df
```

```python
col1='b001100000'
col2='b001101000'
df=merge_col(df,col1,col2)
df.head()
```

# NA 处理

```python
df=df.fillna(0)
df.head()
```

```python
cols = df.columns
droplist=[]
for col in cols:
    # print(df[col].dtype)
    if str(df[col].dtype) != 'float64':
        droplist.append(col)
print(droplist)
```

```python
df=df.drop(droplist,axis=1)
df.head()
```

```python
df=df.fillna(0)
df.head()
```

# 数据标准化

```python
col=list(df.columns)
print(col)
```

```python
for i,name in enumerate(col):
    df[name] = (df[name]-df[name].min())/(df[name].max()-df[name].min())
```
